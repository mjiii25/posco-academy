### Python을 이용한 기술통계 연습 (212쪽)

```python
dat = pd.read_csv('/home/piai/minji/dataset/mycars.csv', encoding = 'euc-kr')

print(dat.shape)
dat.head()
```

```python
dat1 = dat[['automatic', 'model']]

print(dat1.shape)
dat1.head()
```

```python
# 'automatic' 변수의 범주별 개수
count_auto = dat1['automatic'].value_counts().sort_index()


# 누적개수, 확률, 누적확률
cumcnt_auto = np.cumsum(count_auto)
percent_auto = count_auto / sum(count_auto) * 100
cumpct_auto = np.cumsum(percent_auto)
```

```python
count_data_auto = pd.DataFrame({'Count' : count_auto, 'CumCnt' : cumcnt_auto,
                                'Percent' : percent_auto, 'CumPct' : cumpct_auto})
```

```python
count_data_auto
```

### Python을 이용한 확률 계산 실습 (226쪽)

```python
mu = 175
sigma = 5
x = 180

prob = stats.norm.cdf(x, mu, sigma)
print("나의 키가 {0}cm일 경우 내가 위치할 % 범위 : {1:.1f}%".format(x, (1-prob) * 100))
```

### Python을 이용한 확률 계산 실습 (236쪽)

```python
t = 2.0
df = 100

prob = stats.t.cdf(t, df)
print("P(T <= F) : {:.3f}".format(prob))
```

자유도가 100일 때, 2.0에 대한 검정통계량 T의 누적확률은 0.976이다.

### Python을 이용한 확률 계산 실습 (239쪽)

```python
chisq = 3
df = 8

prob = stats.chi2.cdf(chisq, df)
print("P(X <= {0}) : {1:.4f}".format(chisq, prob))
```

자유도가 8일 때, 3에 대한 검정통계량 X^2의 누적확률은 0.0656이다.

### Python을 이용한 확률분포 그리기, 확률 계산 실습 (243쪽)

```python
f = 1.5
dfnum = 10
dfden = 100

prob = stats.f.cdf(x = f, dfn = dfnum, dfd = dfden)
print("P(X <= F): {:.3f}".format(prob))
```

자유도가 각각 10과 100일 때, F = 1.5에 대한 검정통계량 F의 누적확률은 0.850이다.

### Python을 이용한 확률분포 그리기, 확률 계산 실습 (250쪽)

```python
n = 5
for i in range(n+1) :
    prob = stats.binom.pmf(k = i, n = n, p = 0.3)
    print("P(X={0}) = {1:.3f}".format(i, prob))
```

5대의 차량을 임의로 선택했을 때, 불량대수가 각각 0, 1, 2, 3, 4, 5일 확률은 각각 0.168, 0.360, 0.309, 0.132, 0.028, 0.002이다.
