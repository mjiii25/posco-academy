```python
class Set :
    
    
    
    def __init__(self, lst = []) :
        self.lst = []
        for x in lst :
            if x not in self.lst :
                self.lst.append(x)

        
        
    def add(self, elem) :
        if elem not in self.lst :
            self.lst.append(elem)

            
            
    def discard(self, elem) :
        if elem in self.lst :
            self.lst.remove(elem)

            
            
    def clear(self) :
        self.lst = []

        
        
    def __len__(self) :
        return len(self.lst)


    
    
    def __str__(self) :
        return '{' + ','.join(list(map(str, self.lst))) + '}'

    
    
    def __contains__(self, elem) :
        if elem in self.lst :
            return True
        else :
            return False

        
        
    def __le__(self, other) :
        for x in self.lst :
            if x not in other.lst :
                return False
            else :
                return True
            
    
    
    def __ge__(self, other) :
        for x in other.lst :
            if x not in self.lst :
                return False
            else :
                return True
                
                
        
    def __or__(self, other) :
        new = self.lst.copy()
        for x in other.lst :
            if x not in new :
                new.append(x)
        return Set(new)
    

        
        
    def __and__(self, other) :
        new = []
        for x in other.lst :
            if x in self.lst :
                new.append(x)
        return Set(new)
    
    
    
    
    def __sub__(self, other) :
        new = []
        for x in self.lst :
            if x not in other.lst :
                new.append(x)
        return Set(new)
    
    
    def __ior__(self, other) :
        val = self.__or__(other)
        self.lst = val.lst
        return self
    
    def __iand__(self, other) :
        val = self.__and__(other)
        self.lst = val.lst
        return self
    
    def __isub__(self, other) :
        val = self.__sub__(other)
        self.lst = val.lst
        return self
    
    
```

```python
a = Set([1,2,3,4])
b = Set([1,2,3,4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
print()
```

```python
a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print()

a = Set([1, 2, 3])
b = Set([3, 4])
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print()
```
