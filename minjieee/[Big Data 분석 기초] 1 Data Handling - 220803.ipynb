```python
import pandas as pd

df = pd.DataFrame({'ABC' : ['A', 'B', 'C', 'D'],
                  'abc' : ['a', 'b', 'c', 'c'],
                  'group1' : ['G1', 'G2', 'G3', 'G4'],
                  'value1' : [11, 13, 12, 14],
                  'val_mod' : [0.231, 0.197, 0.213, 0.199]})

df
```

**filter 함수**

| Arguments | 기능 |
| ----- | ----- |
| items | 직접 입력한 칼럼 선택 |
| like | 입력값을 포함한 칼럼 선택 |
| regex | regular experssion, 입력값을 포함한 칼럼 선택 |
| axis | 적용 대상 축 |

```python
df.filter(items = ['group1', 'value1'])
```

```python
df.filter(regex = '1', axis = 1)
```

```python
df.filter(regex = '1$', axis = 1)
```

```python
# 'a'를 포함하는 칼럼 출력하기

df.filter(like = 'a', axis = 1)
```

```python
df.filter(regex = '[Aa]', axis = 1)
```

**mask 함수**

| Arguments | 기능 |
| ----- | ----- |
| conds | mask 적용 조건 |
| other | 적용 조건에 해당할 때 처리 |
| inplace | 데이터 변환 후 저장 |

```python
df.mask(df.value1 > 12)
```

```python
df.mask(df.abc == "a")
```

```python
df.mask(cond = df.value1 > 12, other = 999)      # other : NaN을 대체할 값
```
