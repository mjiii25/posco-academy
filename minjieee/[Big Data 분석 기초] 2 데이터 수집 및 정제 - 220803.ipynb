```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
df_raw = pd.read_csv('/home/piai/minji/dataset/FITNESS_결측.csv', encoding = 'euc-kr')
df_raw.head()
```

```python
df_raw.isnull().head()
```

```python
df_raw.isnull().sum()
```

```python
df_raw['GENDER'].fillna('여성', inplace = True)
```

```python
df_raw.head()
```

```python
df_raw.isnull().sum()
```

```python
df_raw.groupby("GENDER")["WEIGHT"].mean()
```

```python
df_raw.groupby("GENDER")['WEIGHT'].agg("mean")
```

```python
df_raw.groupby("GENDER")['WEIGHT'].transform("mean")
```

```python
df_raw['WEIGHT'].fillna(df_raw.groupby('GENDER')['WEIGHT'].transform('mean'), inplace = True)
```

```python
df_raw.isnull().sum()
```

```python
df_raw["RSTPULSE"] >= 100
df_raw.loc[27:29]
```

```python
df_fitness = df_raw[df_raw["RSTPULSE"] < 100]
df_fitness
```

```python
df_fitness.isnull().sum()
```

```python
df_fitness.reset_index(drop = True, inplace = True)
```

```python
df_fitness
```

```python
df_fitness_num = df_fitness.select_dtypes(exclude = "object")
df_fitness_num.head()
```

```python
df_fitness_char = df_fitness.select_dtypes(include = 'object')
df_fitness_char.head()
```

```python
# StandardScaler


scale_std = StandardScaler()
df_scale_std = scale_std.fit_transform(df_fitness_num)

df_scale_std = pd.DataFrame(df_scale_std, columns = df_fitness_num.columns)
df_scale_std.head()
```

```python
df_scale_std.describe().round(3)
```

```python
df_scale_minmax = MinMaxScaler().fit_transform(df_fitness_num)
df_scale_minmax = pd.DataFrame(df_scale_minmax, columns = df_fitness_num.columns)
df_scale_minmax.describe().round(3)
```

```python
df_scaler_robust = RobustScaler().fit_transform(df_fitness_num)
df_scaler_robust = pd.DataFrame(df_scaler_robust, columns = df_fitness_num.columns)
df_scaler_robust.head()
```

```python
df_scaler_robust.describe().round(3)
```

```python
df_rstpulse = pd.DataFrame()

df_rstpulse['Raw'] = df_fitness_num['RSTPULSE']
df_rstpulse['Standard'] = df_scale_std['RSTPULSE']
df_rstpulse['MinMax'] = df_scale_minmax['RSTPULSE']
df_rstpulse['Robust'] = df_scaler_robust['RSTPULSE']
df_rstpulse.head().round(3)
```

```python
df_rstpulse.describe().round(3)
```

```python
df_rstpulse[['Raw', 'Standard', 'MinMax', 'Robust']].hist(figsize = (10,5))
```
