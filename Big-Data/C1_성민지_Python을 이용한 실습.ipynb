### Python을 이용한 데이터 결합 실습 (48쪽)

```python
import pandas as pd


df4 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])

df5 = pd.DataFrame({'E': ['E4', 'E5', 'E6', 'E7'],
                    'F': ['F4', 'F5', 'F6', 'F7'],
                    'G': ['G4', 'G5', 'G6', 'G7'],
                    'H': ['H4', 'H5', 'H6', 'H7']},
                    index=[0, 1, 2, 3])

df6 = pd.DataFrame({'I': ['I8', 'I9', 'I10', 'I11'],
                    'J': ['J8', 'J9', 'J10', 'J11'],
                    'K': ['K8', 'K9', 'K10', 'K11'],
                    'L': ['L8', 'L9', 'L10', 'L11']},
                    index=[0, 1, 2, 3])
```

### Python을 이용한 데이터 변환 실습 (70쪽)

```python
import pandas as pd


fit = pd.read_csv('/home/piai/minji/dataset/FITNESS.csv', encoding = 'euc-kr')
```

```python
print(fit.shape)
fit.head()
```

```python
fit.groupby(["GENDER", "AGEGROUP"], as_index = False)["WEIGHT"].agg("mean").round(3)
```

```python
fit_new = fit.groupby(["GENDER", "AGEGROUP"], as_index = False)["WEIGHT"].agg("mean").round(3)
```

```python
res1 = fit_new.pivot(index = "GENDER", columns = "AGEGROUP")
res1
```

```python
res2 = fit_new.pivot(index = "AGEGROUP", columns = "GENDER")
res2
```
